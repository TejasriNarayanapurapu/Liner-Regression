```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# **Data Collection**

# Create a dictionary with data
# **Data Collection**

# Create a dictionary with data
data = {
    'Maths': [17,72,97,8,32,15,63,97,57,60,83,48,100,26,12,62,3,49,55,77,97,98],
     'Physics': [27,82,18,42,25,73,67,70,93,58,36,22,72,13,59,65,87,10,67,44,39,85]
}

# Create a DataFrame
import pandas as pd
df = pd.DataFrame(data)
df
# Create a DataFrame
df = pd.DataFrame(data)
df
```

```python
# Save the DataFrame to a CSV file
df.to_csv('Student.csv', index=False)
```

```python
# **Data Exploration**

# Load dataset
data = pd.read_csv('/content/Student.csv')
# data

# Features and target variable
X = df[['Maths']]
y = df['Physics']

print(type(X))
print(type(y))
y
```

```python
# **Data Splitting**


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# **Model Building**

# Initialize the model
model = LinearRegression()
```

```python
 #**Model Training**


# Train the model
model.fit(X_train, y_train)


# Make predictions
y_pred = model.predict(X_test)

print(y_pred)
print(y_test)
```

```python
# Plotting the regression line and data points
plt.scatter(X, y, color='skyblue')  # Data points
#plt.plot(X, model.predict(X), color='red')  # Regression line
plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Linear Regression - Maths vs Physics')
plt.show()
```

```python
# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
